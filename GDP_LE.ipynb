# ANALYSIS OF GDP AND LIFE EXPECTANCY

The purpose of this project is to analyse the changes in GDP and life expectancy over 15 years (2000 - 2015) in 6 countries (Chile, China, Germany, Mexico, USA and Zimbabwe) and inspect the relationship between GDP of a country and life expectancy of its citizens.

**Data sources**

- GDP Source: [World Bank](https://data.worldbank.org/indicator/NY.GDP.MKTP.CD) national accounts data, and OECD National Accounts data files.

- Life expectancy Data Source: [World Health Organization](http://apps.who.int/gho/data/node.main.688)

```python
%config IPCompleter.greedy=True
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('all_data.csv')
```

```python
df.rename(columns={'Country': 'country', 'Year': 'year', 'Life expectancy at birth (years)': 'life_exp', 'GDP': 'gdp'}, inplace=True)
```

## Change in GDP over time

```python
sns.lineplot(x=df.year, y=df.gdp, hue=df.country, palette='colorblind')
plt.title("GDP per country")
plt.ylabel("GDP")
plt.xlabel("Year")
plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.show()
plt.clf()
```

```python
graph1 = sns.FacetGrid(df, col='country', hue='country', col_wrap=3, sharey = False, sharex = False, palette='colorblind')
graph1.map(sns.lineplot, 'year', 'gdp').set_axis_labels("Year", "GDP in Trillions of U.S. Dollars")
graph1.set_titles(col_template="{col_name}")
graph1.fig.subplots_adjust(top=0.9)
graph1.fig.suptitle('GDP per country')
```

GDP increased between 2000 and 2015 in all countries in the dataset. USA and China saw the greatest gains.

## Change in life expectancy over time

```python
comparison = df[(df.year == 2000) | (df.year == 2015)]
ax = sns.barplot(x='country', y='life_exp', data=comparison, hue='year', palette='colorblind')
ax.set_xticklabels(['Chile', 'China', 'Germany', 'Mexico', 'USA', 'Zimbabwe'])
plt.title("...")
plt.ylabel("Life expectancy at birth (years)")
plt.xlabel("Country")
plt.show()
plt.clf()
```

Life expectancy increased in all tracked countries. Zimbabwe saw the sharpest increase, however it remains behind compared to the rest of the analysed group.

```python
sns.lineplot(x=df.year, y=df.life_exp, hue=df.country, palette='colorblind')
plt.title("Life expectancy per country")
plt.ylabel("Life expectancy at birth (years)")
plt.xlabel("Year")
plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.show()
plt.show()
plt.clf()
```

```python
graph2 = sns.FacetGrid(df, col='country', hue='country', col_wrap=3, sharey = False, sharex = False, palette='colorblind')
graph2.map(sns.lineplot, 'year', 'life_exp').set_axis_labels("Year", "Life expectancy at birth (years)")
graph2.set_titles(col_template="{col_name}")
graph2.fig.subplots_adjust(top=0.9)
graph2.fig.suptitle('Life expectancy per country')
```

## Relationship between GDP and life expectancy

```python
graph3 = sns.FacetGrid(df, col='country', hue='country', col_wrap=3, sharey = False, sharex = False, palette='colorblind')
graph3.map(sns.scatterplot, 'life_exp', 'gdp').set_axis_labels("Life expectancy at birth (years)", "GDP in Trillions of U.S. Dollars")
graph3.set_titles(col_template="{col_name}")
graph3.fig.subplots_adjust(top=0.9)
graph3.fig.suptitle('GDP vs Life expectancy')
```

Scatter plots above explore the relationship between GDP and life expectancy at birth for each country in the dataset. Most countries have linear relationship between GDP and life expectancy. All of them exhibit positive correlation.

## Conclusions

- GDP increased for all countries in the dataset, with China and USA dominating with sharpest increases.
- Life expectancy increased over time in all countries in the dataset, with Zimbabwe recording the greatest increase. Despite that, life expectancy in Zimbabwe lags behind the rest of the group.
- There is a positive correlation between GDP and life expectancy for countries in the dataset.
